```python
# Import necessary libraries/modules
import vertica_python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import os
```

```python
# Set up the Vertica Connection Information
username = os.getenv('VERTICA_USERNAME')
password = os.getenv('VERTICA_PASSWORD')

conn_info = {'host': 'warehouse.analytics.edx.org',
             'port': 5433,
             'user': username,
             'password': password,
             'database': 'warehouse',
             'read_timeout': 600,
             'unicode_error': 'strict',
             'ssl': False
            }

connection = vertica_python.connect(**conn_info)
cur = connection.cursor('dict')
```

```python
query = """
DROP TABLE IF EXISTS ed_services.relative_to_start;
CREATE TABLE IF NOT EXISTS ed_services.relative_to_start AS
SELECT
	a.course_id,
	DATEDIFF(day, date, course_start_date) AS days_from_start,
	cnt_enrolls
FROM
	business_intelligence.course_stats_time a
LEFT JOIN
	business_intelligence.course_master b
ON
	a.course_id = b.course_id
WHERE
    course_start_date < CURRENT_DATE();
    
DROP TABLE IF EXISTS ed_services.relative_to_end;
CREATE TABLE IF NOT EXISTS ed_services.relative_to_end AS
SELECT
	a.course_id,
	DATEDIFF(day, date, course_end_date) AS days_from_end,
	cnt_enrolls
FROM
	business_intelligence.course_stats_time a
LEFT JOIN
	business_intelligence.course_master b
ON
	a.course_id = b.course_id
WHERE
    course_end_date < CURRENT_DATE();    
"""

start_query = """
SELECT
	days_from_start,
	SUM(cnt_enrolls) AS enrollments
FROM
	ed_services.relative_to_start
GROUP BY
	days_from_start
HAVING
	SUM(cnt_enrolls) > 0;
"""

end_query = """
SELECT
	days_from_end,
	SUM(cnt_enrolls) AS enrollments
FROM
	ed_services.relative_to_end
GROUP BY
	days_from_end
HAVING
	SUM(cnt_enrolls) > 0;
"""

# Execute the Query from above
cur.execute(query)

cur.execute(start_query)
# Dump the query results into a Pandas DataFrame
df_start = pd.DataFrame(cur.fetchall())

cur.execute(end_query)

df_end = pd.DataFrame(cur.fetchall())
```

```python
sns.regplot(x=df_start['days_from_start'], y=df_start['enrollments'], fit_reg=False)
```

```python
sns.regplot(x=df_end['days_from_end'], y=df_end['enrollments'], fit_reg=False)
```

```python
query = """
SELECT
	DATEDIFF(day, course_announcement_date, course_start_date) AS days_between_announce_start,
    pacing_type,
    level_type,
	SUM(sum_enrolls) AS enrollments
FROM
	business_intelligence.course_stats_summary a
LEFT JOIN
	business_intelligence.course_master b
ON
	a.course_id = b.course_id
GROUP BY
	1, 2, 3
HAVING
    DATEDIFF(day, course_announcement_date, course_start_date) > 0;
"""

cur.execute(query)

df = pd.DataFrame(cur.fetchall())
```

```python
sns.regplot(x=df['days_between_announce_start'], y=df['enrollments'], fit_reg=False)
```

```python
ax = sns.boxplot(x=df['days_between_announce_start'], y=df['pacing_type'])
# instructor paced courses tend to have longer lead time which is interesting
```

```python
ax = sns.boxplot(x=df['days_between_announce_start'], y=df['level_type'])
# Not much difference here, slightly longer lead time for Introductory courses
```

```python
ax = sns.boxplot(x=df['days_between_announce_start'], y=df['level_type'], hue=df['pacing_type'])
# same trend here across all levels, instructor paced seems to have longer lead time
```

```python
df['days_bucket'] = 'Quartile'

df['days_bucket'][df['days_between_announce_start'] >= df['days_between_announce_start'].quantile(.75)] = 'Quartile 4'
df['days_bucket'][df['days_between_announce_start'] < df['days_between_announce_start'].quantile(.75)] = 'Quartile 3'
df['days_bucket'][df['days_between_announce_start'] < df['days_between_announce_start'].quantile(.5)] = 'Quartile 2'
df['days_bucket'][df['days_between_announce_start'] < df['days_between_announce_start'].quantile(.25)] = 'Quartile 1'
```

```python
ax = sns.boxplot(x=df['enrollments'], y=df['days_bucket'], hue=df['level_type'], order=['Quartile 4', 'Quartile 3', 'Quartile 2', 'Quartile 1'])
# so we see a rise in enrollments when the announcement to start window is small
```

```python
query = """
SELECT
	DATEDIFF(day, course_start_date, course_end_date) AS course_length,
    pacing_type,
    level_type,
	SUM(sum_enrolls) AS enrollments
FROM
	business_intelligence.course_stats_summary a
LEFT JOIN
	business_intelligence.course_master b
ON
	a.course_id = b.course_id
GROUP BY
	1, 2, 3
HAVING
    DATEDIFF(day, course_start_date, course_end_date) > 0;
"""

cur.execute(query)

df = pd.DataFrame(cur.fetchall())
```

```python
ax = sns.boxplot(x=df['course_length'], y=df['pacing_type'])
ax.set(xlim=(0, 1000))
# self paced courses are obviously longer
```

```python
ax = sns.boxplot(x=df['course_length'], y=df['level_type'])
ax.set(xlim=(0, 1000))
# Easier courses tend to be longer which is interesting since I woud've assumed the opposite
# Might be worth looking at sample sizes here, we might just have a lot more introductory courses than any others
```

```python
ax = sns.boxplot(x=df['course_length'], y=df['level_type'], hue=df['pacing_type'])
ax.set(xlim=(0, 1000))
# Same trend here, self paced courses are open much longer and we see longer open times for easier level courses
```

```python
df['days_bucket'] = 'Quartile'

df['days_bucket'][df['course_length'] >= df['course_length'].quantile(.75)] = 'Quartile 4'
df['days_bucket'][df['course_length'] < df['course_length'].quantile(.75)] = 'Quartile 3'
df['days_bucket'][df['course_length'] < df['course_length'].quantile(.5)] = 'Quartile 2'
df['days_bucket'][df['course_length'] < df['course_length'].quantile(.25)] = 'Quartile 1'
```

```python
query = """
SELECT
	days_to_verification,
	SUM(cnt_verifications) AS verifications
FROM
	(
		SELECT
			a.course_id,
			DATEDIFF(day, date, course_verification_end_date) AS days_to_verification,
			cnt_verifications
		FROM
			business_intelligence.course_stats_time a
		LEFT JOIN 
			business_intelligence.course_master b
		ON
			a.course_id = b.course_id
        WHERE
            course_verification_end_date IS NOT NULL
        AND
            course_end_date < CURRENT_DATE()
	) a
GROUP BY
	1
HAVING
	SUM(cnt_verifications) > 0;
"""

cur.execute(query)

df_verifications = pd.DataFrame(cur.fetchall())
```

```python
sns.regplot(x=df_verifications['days_to_verification'], y=df_verifications['verifications'], fit_reg=False)
```

```python
query = """
SELECT
	days_from_start,
	SUM(cnt_enrolls) AS enrollments
FROM
	(
		SELECT
			a.course_id,
			DATEDIFF(day, date, course_start_date) AS days_from_start,
			cnt_enrolls
		FROM
			business_intelligence.course_master a
		LEFT JOIN
			business_intelligence.course_stats_time b
		ON
			a.course_id = b.course_id
		WHERE
			course_verification_end_date IS NOT NULL
		AND
			a.course_id NOT IN ('course-v1:ASUx+AST111+3T2015', 'course-v1:MITx+ENx+2016_3T')
		AND
			pacing_type = 'instructor_paced'
		AND
			course_end_date < CURRENT_DATE()
		AND
			course_start_date IS NOT NULL
	) a
GROUP BY
	1
HAVING
	SUM(cnt_enrolls) > 0;
"""

cur.execute(query)
instructor_paced_df = pd.DataFrame(cur.fetchall())
```

```python
ax = sns.regplot(x=instructor_paced_df['days_from_start'], y=instructor_paced_df['enrollments'], fit_reg=False)
ax.set(xlim=(-1000, 1000))
```

```python
query = """
SELECT
	days_from_start,
	SUM(cnt_enrolls) AS enrollments
FROM
	(
		SELECT
			a.course_id,
			DATEDIFF(day, date, course_start_date) AS days_from_start,
			cnt_enrolls
		FROM
			business_intelligence.course_master a
		LEFT JOIN
			business_intelligence.course_stats_time b
		ON
			a.course_id = b.course_id
		WHERE
			course_verification_end_date IS NOT NULL
		AND
			a.course_id NOT IN ('course-v1:ASUx+AST111+3T2015', 'course-v1:MITx+ENx+2016_3T')
		AND
			pacing_type = 'self_paced'
		AND
			course_start_date IS NOT NULL
        AND
            course_start_date + 180 < CURRENT_DATE()
	) a
GROUP BY
	1
HAVING
	SUM(cnt_enrolls) > 0;
"""

cur.execute(query)
self_paced_df = pd.DataFrame(cur.fetchall())
```

```python
sns.regplot(x=self_paced_df['days_from_start'], y=self_paced_df['enrollments'], fit_reg=False)
```

```python
query = """
SELECT
	days_between_enroll_to_verify,
	COUNT(*) AS verified_learners
FROM
	(
		SELECT
			user_id,
			course_id,
			DATEDIFF(day, first_enrollment_time, first_verified_enrollment_time) AS days_between_enroll_to_verify
		FROM
			production.d_user_course
		WHERE
			first_verified_enrollment_time IS NOT NULL
	) a
GROUP BY
	1;
"""

cur.execute(query)

df = pd.DataFrame(cur.fetchall())
```

```python
sns.regplot(x=df['days_between_enroll_to_verify'], y=df['verified_learners'], fit_reg=False)
# so most people are enrolling and verifying on the same day
```

```python
query = """
SELECT
	days_from_start,
	COUNT(*) AS same_day_verified_users
FROM
	(
		SELECT
			user_id,
			a.course_id,
			first_enrollment_time,
			first_verified_enrollment_time,
			course_start_date,
			DATEDIFF(day, first_enrollment_time, course_start_date) AS days_from_start
		FROM
			production.d_user_course a
		LEFT JOIN
			(
				SELECT
					*
				FROM
					business_intelligence.course_master 
				WHERE
					course_verification_end_date IS NOT NULL
				AND
					course_id NOT IN ('course-v1:ASUx+AST111+3T2015', 'course-v1:MITx+ENx+2016_3T')
			) b
		ON
			a.course_id = b.course_id
		WHERE
			first_verified_enrollment_time IS NOT NULL
		AND
			DATEDIFF(day, first_enrollment_time, first_verified_enrollment_time) = 0
	) a
GROUP BY
	1;
"""

cur.execute(query)
df = pd.DataFrame(cur.fetchall())
```

```python
sns.regplot(x=df['days_from_start'], y=df['same_day_verified_users'], fit_reg=False)
```

```python
query = """
        SELECT
			user_id,
			course_id,
			DATEDIFF(day, first_enrollment_time, first_verified_enrollment_time) AS days_between_enroll_to_verify
		FROM
			production.d_user_course
		WHERE
			first_verified_enrollment_time IS NOT NULL
"""

cur.execute(query)

df = pd.DataFrame(cur.fetchall())
```

```python
# from the above query
# 42% of verifications come on the same day as enrollment
# 54% within 1 week
# 61% within 2 weeks
# 67% within 3 weeks
# 72% within 4 weeks
# 85% within 8 weeks
# 92% within 12 weeks
# 95% within 16 weeks
# 97 within 20 weeks
```

```python
df['day_buckets'] = 'Null'

df['day_buckets'][df['days_between_enroll_to_verify'] <= 140] = '20 weeks'
df['day_buckets'][df['days_between_enroll_to_verify'] <= 112] = '16 weeks'
df['day_buckets'][df['days_between_enroll_to_verify'] <= 84] = '12 weeks'
df['day_buckets'][df['days_between_enroll_to_verify'] <= 56] = '8 weeks'
df['day_buckets'][df['days_between_enroll_to_verify'] <= 28] = '4 weeks'
df['day_buckets'][df['days_between_enroll_to_verify'] <= 21] = '3 weeks'
df['day_buckets'][df['days_between_enroll_to_verify'] <= 14] = '2 weeks'
df['day_buckets'][df['days_between_enroll_to_verify'] <= 7] = '1 week'

df['day_buckets'][df['days_between_enroll_to_verify'] >140] = '> 20 weeks'
```

```python
days_between_enroll_to_verify = pd.DataFrame(df.groupby(['day_buckets'])['user_id'].count())
days_between_enroll_to_verify
```

```python
query = """
SELECT
	DATEDIFF(day, first_enrollment_time, course_start_date) AS days_between_start_enroll,
	COUNT(*) AS users
FROM
	(
		SELECT
			user_id,
			course_id,
			DATEDIFF(day, first_enrollment_time, first_verified_enrollment_time) AS days_between_enroll_to_verify,
			first_enrollment_time
		FROM
			production.d_user_course
		WHERE
			first_verified_enrollment_time IS NOT NULL
		AND
			DATEDIFF(day, first_enrollment_time, first_verified_enrollment_time) <= 28
	) a
LEFT JOIN
	ed_services.course_master b
ON
	a.course_id = b.course_id
WHERE
	course_start_date < CURRENT_DATE()
GROUP BY
	1;
"""

cur.execute(query)

df = pd.DataFrame(cur.fetchall())
```

```python
ax = sns.regplot(x=df['days_between_start_enroll'], y=df['users'], fit_reg=False)
ax.set(ylim=(0, 8000))
#% of users from certain demographic
```
