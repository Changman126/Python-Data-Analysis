# Logistic Regression

Statistical technique capable of predicting a binary outcome. Basically a linear model for classification rather than regression. Also known as logit regression, maximum-entropy classification, or the log linear classifier.

Why use dummy columns and exclude a baseline from fit? To prevent multicollinearity or the dummy variable trap caused by including a dummy variable for every single category.

Multicollinearity = a scenario in which two or more variables are highly correlated; basically one variable can be predicted from the other.

Example: Using gender (male/female) as a dummy column for each variable is redundant (of male is 0, female is 1, and vice versa). This causes multicollinearity and thus the reason why we drop one category if it is inherently defined by another category.

```python
# Instantiate a logistic regression model and fit with X and y
model = LogisticRegression()
model = model.fit(X, y)

# check the accuracy on the training set
model.score(X, y)

# runs logistic regression on the entire data set and shows how accurate it is
```

```python
# percentage of courses above average
# means you can obtain 50% accuracy by always predicting no
# worse than the null error rate...
y.mean()
```

```python
# examine the coefficients
pd.DataFrame(zip(X.columns, np.transpose(model.coef_)))

# increase in Architecture corresponds to an increase in the likelihood
# of being an above average course
```

```python
# Evaluate the model by splitting into training and test sets
X_train, X_test, y_train, y_test = train_test_split(course_df[feature_cols], course_df['above_average_enrollments'], test_size=0.20, random_state=42)

model2 = LogisticRegression()
model2.fit(X_train, y_train)

# predict performance [y] for the test set based on the training set
prediction = model2.predict(X_test)

# generate class probabilities
probs = model2.predict_proba(X_test)

print metrics.accuracy_score(y_test, prediction)
print metrics.roc_auc_score(y_test, probs[:, 1])

# The accuracy is 67%, which is worse than before splitting the data set
```

# What about Decision Trees?

Tree based learning algorithms are considered to be one of the best and mostly used supervised learning methods. Tree based methods empower predictive models with high accuracy, stability, and ease of interpretation. 

Unlike linear models, they map non-linear relationships quite well and are adaptable at solving any kind of problem at hand (classification or regression)

# What is a Decision Tree? How does it work?

Decision tree is a type of supervised learning algorithm (having a pre-defined target variable) that is mostly used in classification problems. It works for both categorical and continuous input and output variables.

In this technique, we split the population or sample into two or more homogeneous sets (or sub-populations) based on the most significant splitter/differentiator in input variables.

# Let's try a Random Forest Classifier

# What's the difference between a Random Forest and Decision Tree

The superficial answer is that a Random Forest is a collection of Decision Trees

Decision trees are more susceptive to overfitting, by memorizing the training set similar to the way a person might memorize an Eye Chart.

The point of Random Forest is to prevent overfitting by creating random subsets of the features and building smaller (shallow) trees using the subsets and then combining it into subtrees.

The downside to Random Fores is that it can be slow if you have a sinngle process.

# Random Forests
Random forest belongs to the ensemble methods learning class which involes the combination of several models to solve a single prediction problem. It works by generating multiple classifiers/models which learn and make predictions independently. Those predictions are then combined into a single prediction that should be good or better than the prediction made by any one classifier.

When you make a prediction, the new observation gets pushed down each decision tree and assigned a predicted value/label. Once each of the trees in the forest have reported its predicted value/label, the predictions are tallied up and the mode vote of all trees is returned as the final prediction.

Simply, the 99.9% of trees that are irrelevant make predictions that are all over the map and cancel each other out. The predictions of the minority of trees that are good top that noise and yield a good prediction.

# Why should you use it?
It's the leatherman of learning methods. You can throw basically anything at it and it will do a serviceable job. It does a particularly good job of estimating inferred transformations and doesn't require much tuning.

# Best Use Case
Feature Selection

You can examine which variables are working best/worst in each tree.

When a certain tree uses one variable and another doesn't, you cna compare the value lost or gained from the inclusion/exclusion of that variable. The good random forest implementations are goig to do that for you, so all you need to do is know which method or variable to look at.

Classification

Can be used to make predictions for categories with multiple possible values and can be calibrated to output probabilities as well. 

Regression

Does really well learning on categorical variables or a mixture of categorical and real variables. Categorical variables with high cardinality (# of possible values) can be tricky, so having something like this in your back pocket can come in quite useful.

# Danger

Overfitting

Random forest can be prone to overfitting, especially when working with relatively small datasets. You should be suspicious if your model is making "too good" of predictions on our test set.

One way to avoid overfitting is to only use really relevant features in your model. It can make it a lot easier
